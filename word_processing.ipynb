```python
import numpy as np
from tqdm.notebook import tqdm
from sentence_transformers import SentenceTransformer
```

```python
with open('word_lists/wiki-100k.txt') as f:
    words = np.array(f.read().splitlines())
len(words)
```

```python
words = [w for w in words if not w.startswith('#')]
len(words)
```

```python
def f7(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]

words = [w.upper() for w in words]
words = f7(words)
len(words)
```

```python
words
```

```python
# Initialize the embedding model.
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
print(embedding_model)
```

```python
with open("words.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(words))
```

```python
candidate_embeddings = np.array([embedding_model.encode(word) for word in tqdm(words)])
```

```python
np.save("clue_candidate_embeddings.npy", candidate_embeddings)
```
